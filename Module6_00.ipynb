## 2. Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables.



```python
# Sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Scatter plot
plt.scatter(x, y, label='Data Points'); plt.xlabel('X-axis'); plt.ylabel('Y-axis'); plt.title('Scatter Plot'); plt.legend(); plt.show()


```

## 3. Line Plot
A line plot is often used to display trends over time.

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line plot
plt.plot(x, y, label='sin(x)'); plt.xlabel('X-axis'); plt.ylabel('Y-axis'); plt.title('Line Plot of sin(x)'); plt.legend(); plt.show()


```

## 6. Time Series Plot
A time series plot is used to visualize data over time.

```python
# Generate a time series dataset
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
data = np.random.randn(100).cumsum()

# Time series plot
plt.plot(dates, data); plt.title('Time Series Plot'); plt.xlabel('Date'); plt.ylabel('Cumulative Value'); plt.grid(True); plt.show()




```

## 8. Combining Multiple Plots in One Figure
You can also combine multiple plots into one figure using plt.subplot().

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_scatter = np.random.rand(10)
y_box = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 2, 4]

# Set up the subplot grid
plt.figure(figsize=(10, 8))

# Scatter plot
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.scatter(x[:10], y_scatter, color='blue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Line plot
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.plot(x, y, color='red')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Sine Value')

# Box plot
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.boxplot(y_box)
plt.title('Box Plot')
plt.ylabel('Value')

# Bar graph
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.bar(categories, values, color='green')
plt.title('Bar Graph')
plt.ylabel('Value')

# Show all plots
plt.tight_layout()  # Adjusts subplot parameters for better spacing
plt.show()


```
